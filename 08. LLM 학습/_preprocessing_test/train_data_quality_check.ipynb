```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd

# 예제 데이터
texts = [
    "오늘 날씨가 좋다",
    "오늘의 날씨는 참 좋다",
    "주식 시장이 상승했다",
    "오늘 날씨가 좋다"  # 중복
]

# n-gram 기반 유사도 (여기선 2-gram 사용)
vectorizer = CountVectorizer(ngram_range=(2,2)).fit(texts)
X = vectorizer.transform(texts)

# Cosine similarity로 유사도 측정

similarity = cosine_similarity(X)

# 중복 탐지 (예: 0.8 이상이면 유사하다고 판단)
dup_pairs = [(i,j) for i in range(len(texts)) for j in range(i+1, len(texts)) if similarity[i,j] > 0.8]

print("중복 후보:", dup_pairs)
```
