```python
def solution(m, n, h, w, drops):
    rain_map = [[0] * n for _ in range(m)]

    for time, drop in enumerate(drops, start=1):
        r, c = drop
        rain_map[r][c] = time
    best_time = -1
    answer = [0, 0]

    for i in range(m - h + 1):
        for j in range(n - w + 1):

            first_rain_time = float("inf")

            for r in range(i, i + h):
                for c in range(j, j + w):
                    if rain_map[r][c] != 0:
                        first_rain_time = min(first_rain_time, rain_map[r][c])

            if first_rain_time > best_time:
                best_time = first_rain_time
                answer = [i, j]

    return answer
```

```python
from collections import deque

def solution(m, n, h, w, drops):
    INF = len(drops) + 1

    # 1. 비가 오는 시간을 기록
    # INF는 비가 오지 않는 칸을 의미
    rain_map = [[INF] * n for _ in range(m)]
    print(rain_map)
    
    for time, drop in enumerate(drops, start=1):
        r, c = drop
    
        # 같은 칸에 비가 여러 번 올 수도 있으니 최초 시간만 기록
        if rain_map[r][c] == INF:
            rain_map[r][c] = time
    print(rain_map)
    # 2. 각 행마다 가로 w칸 구간의 최솟값 구하기
    # row_min[i][j] = i번째 행에서 j ~ j+w-1 구간의 최솟값
    row_min = [[0] * (n - w + 1) for _ in range(m)]

    for i in range(m):
        dq = deque()

        for c in range(n):
            # 현재 값보다 큰 값들은 뒤에서 제거
            while dq and rain_map[i][dq[-1]] >= rain_map[i][c]:
                dq.pop()

            dq.append(c)

            # 윈도우 범위를 벗어난 인덱스 제거
            if dq[0] <= c - w:
                dq.popleft()

            # c가 w-1 이상이면 하나의 가로 윈도우 완성
            if c >= w - 1:
                start_col = c - w + 1
                row_min[i][start_col] = rain_map[i][dq[0]]

    # 3. 세로 h칸에 대해 다시 최솟값 구하기
    # 즉, h x w 직사각형 영역의 최솟값을 구함
    best_time = -1
    answer = [0, 0]

    for j in range(n - w + 1):
        dq = deque()

        for r in range(m):
            # 현재 값보다 큰 값들은 뒤에서 제거
            while dq and row_min[dq[-1]][j] >= row_min[r][j]:
                dq.pop()

            dq.append(r)

            # 윈도우 범위를 벗어난 인덱스 제거
            if dq[0] <= r - h:
                dq.popleft()

            # r이 h-1 이상이면 하나의 h x w 영역 완성
            if r >= h - 1:
                start_row = r - h + 1
                first_rain_time = row_min[dq[0]][j]

                # first_rain_time이 클수록 늦게 비를 맞음
                # INF면 아예 비를 맞지 않는 영역
                if first_rain_time > best_time:
                    best_time = first_rain_time
                    answer = [start_row, j]

    return answer
```
