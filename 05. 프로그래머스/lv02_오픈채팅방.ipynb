```python
record = ["Enter uid1234 Muzi", 
          "Enter uid4567 Prodo",
          "Leave uid1234",
          "Enter uid1234 Prodo",
          "Change uid4567 Ryan"]
```

```python
nickname_map = {}

for rec in record : 
    split_data = rec.split(' ')
    status = split_data[0]
    _id = split_data[1]
    try : 
        nickname = split_data[2]
    except Exception as e : 
        nickname = None
    
    if nickname != None : 
        nickname_map[_id] = nickname

answer = []

for rec in record : 
    tmp_ans = ''
    split_data = rec.split(' ')
    status = split_data[0]
    _id = split_data[1]

    nickname = nickname_map[_id]

    tmp_ans += nickname + '님이 '
    if status == "Enter" : 
        tmp_ans += "들어왔습니다."
    elif status == "Leave" : 
        tmp_ans += '나갔습니다.'
    else : continue
    answer.append(tmp_ans)
    




    
```
