```python
def solution(x, y, n):
    answer = 0
    if x == y:
        return 0
    current_values = [x for i in range(3)]
    while True : 
        answer += 1
        x1 = current_values[0] * 2
        x2 = current_values[1] * 3
        x3 = current_values[2] + n
        current_values = [x1, x2, x3]
        if x1 == y or x2 == y or x3 == y : 
            break

        elif x1 > y and x2 > y and x3 > y : 
            return -1

    return answer
```

```python
solution(x, y, n)
```

```python
from collections import deque

def solution(x, y, n):
    if x == y:
        return 0

    q = deque([(x, 0)])
    visited = set([x])

    while q:
        value, cnt = q.popleft()

        for next_val in (value * 2, value * 3, value + n):
            if next_val == y:
                return cnt + 1
            if next_val < y and next_val not in visited:  # 범위 내 & 미방문
                visited.add(next_val)
                q.append((next_val, cnt + 1))

    return -1  # 도달 불가
```
