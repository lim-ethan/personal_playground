```python
### 튜플 대소관계
# (0, 0) < (0, 1)    # True
# (0, 1) < (0, 0)    # False
# (1, 2) < (1, 3)    # True
# (2, 0) < (1, 5)    # False
```


이동	조건	저장된 path <br>
(0,0) → (0,1)	(0,0) < (0,1) → True	((0,0), (0,1))<br>
(0,1) → (0,0)	(0,1) < (0,0) → False	((0,0), (0,1))


```python
def solution(dirs):
    dict_dir = {
        'U': (0, 1),
        'D': (0, -1),
        'L': (-1, 0),
        'R': (1, 0)
    }

    x, y = 0, 0
    visited = set()
    answer = 0

    for d in dirs:
        dx, dy = dict_dir[d]
        nx, ny = x + dx, y + dy

        if -5 <= nx <= 5 and -5 <= ny <= 5:
            path = ((x, y), (nx, ny)) if (x, y) < (nx, ny) else ((nx, ny), (x, y)) ## 만약 (0, 0) → (0, 1) 이동 시 ((0, 1), (0, 0))로 표기하여 visited 변수에 추가
            print('=' * 150)
            print('ㅁ PATH : ', path)
            print('ㅁ VISITED : ', visited)
            if path not in visited:
                visited.add(path)
                answer += 1
            x, y = nx, ny

    return answer
```

```python
dirs = "LULLLLLLU"
solution(dirs)
```
