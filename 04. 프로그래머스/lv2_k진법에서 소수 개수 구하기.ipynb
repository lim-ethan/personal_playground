```python
import collections

def convert_to_base(n, base):
    digits = '0123456789'
    q = collections.deque()

    if n == 0:
        return '0'

    while n > 0:
        q.appendleft(digits[n % base])
        n = n // base
    return ''.join(q)
```

```python
def is_prime(x):
    if x < 2:
        return False
    for i in range(2, int(x ** 0.5) + 1):
        if x % i == 0:
            return False
    return True
```

```python
def solution(n, k):
    conv_result = convert_to_base(n, k)
    parts = conv_result.split('0')
    answer = 0

    for conv in parts:
        if not conv:
            continue
        if is_prime(int(conv)):
            answer += 1
    return answer
```

```python
n, k = 	110011, 10

solution(n, k)
```
