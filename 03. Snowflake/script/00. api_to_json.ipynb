```python
api_key = "" # gonggong_seoul_api_key

api_endpoints = {
    "MEGA_WORK_POPULATION": f"http://openapi.seoul.go.kr:8088/{api_key}/json/VwsmMegaWrcPopltnW/1/50/",
    "DISTRICT_APARTMENT": f"http://openapi.seoul.go.kr:8088/{api_key}/json/VwsmSignguNcmCnsmpW/1/700/"
}


local_url = {
    "DISTRICT_INCOME_CONSUMPTION": f"http://openapi.seoul.go.kr:8088/{api_key}/json/VwsmSignguAptW/1/700/"
    }
```

```python
import requests
import json
import os


for table_name, url in local_url.items():
    os.makedirs(table_name, exist_ok=True)
    cnt = 0

    response = requests.get(url)
    data = response.json()
    main_key = list(data.keys())[0]
    rows = data[main_key]['row']
    for row in rows : 
        with open(f'{table_name}/row_number_{str(cnt).zfill(3)}.json', 'w', encoding='utf-8') as f : 
            json.dump(row, f, ensure_ascii=False)
        cnt += 1
```
